```python
import tkinter as tk
from tkinter import scrolledtext
from predict import FakeNewsPredictor

class FakeNewsApp:
    def __init__(self, root):
        self.root = root
        self.root.title("📰 AI Fake News Detector")
        self.root.geometry("700x500")
        self.root.configure(bg="#f0f4f7")

        self.predictor = FakeNewsPredictor()
        self.create_frames()
        self.show_frame("home")

    def create_frames(self):
        # Home Frame
        self.home_frame = tk.Frame(self.root, bg="#f0f4f7")
        tk.Label(self.home_frame, text="Welcome to Fake News Detection",
                 font=("Helvetica", 24, "bold"), fg="red", bg="#f0f4f7", pady=40).pack()
        tk.Button(self.home_frame, text="Go to Classifier", command=lambda: self.show_frame("classifier"),
                  font=("Arial", 14), bg="#0a9396", fg="white", padx=20, pady=10).pack(pady=10)
        tk.Button(self.home_frame, text="Exit", command=self.root.destroy,
                  font=("Arial", 12), bg="grey", fg="white", padx=15, pady=5).pack(pady=10)

        # Classifier Frame
        self.classifier_frame = tk.Frame(self.root, bg="#f0f4f7")
        tk.Label(self.classifier_frame, text="Fake News Detector", font=("Helvetica", 20, "bold"),
                 bg="#005f73", fg="white", pady=10).pack(fill="x")
        tk.Label(self.classifier_frame, text="Paste or type the news content below:",
                 font=("Arial", 13), bg="#f0f4f7").pack(pady=10)
        self.input_text = scrolledtext.ScrolledText(self.classifier_frame, height=10, width=80,
                                                    font=("Arial", 12), wrap="word")
        self.input_text.pack(padx=20)
        tk.Button(self.classifier_frame, text="Classify News", command=self.classify_news,
                  font=("Arial", 12, "bold"), bg="#0a9396", fg="white", padx=10, pady=5).pack(pady=15)
        tk.Button(self.classifier_frame, text="⬅ Back", command=lambda: self.show_frame("home"),
                  font=("Arial", 12), bg="#ccc", padx=10, pady=5).pack(pady=10)

    def show_frame(self, name):
        self.home_frame.pack_forget()
        self.classifier_frame.pack_forget()
        if name == "home":
            self.home_frame.pack(fill="both", expand=True)
        elif name == "classifier":
            self.classifier_frame.pack(fill="both", expand=True)

    def classify_news(self):
        news = self.input_text.get("1.0", "end-1c").strip()
        result = self.predictor.predict(news)
        status = result.pop('Status', 'Known')
        is_real = list(result.values()).count("Not Fake News") > list(result.values()).count("Fake News")

        if status == 'EmptyInput':
            self.show_popup("⚠ Please enter some news content.", "#ffcc00")
        elif status == 'Unrecognized':
            self.show_popup("🔍 This news is not found in our training database.", "#fca311")
        else:
            message = "✅ This news appears to be REAL and trustworthy." if is_real else "🚨 This news appears to be FAKE or misleading."
            color = "#2a9d8f" if is_real else "#e63946"
            self.show_popup(message, color)

    def show_popup(self, message, bg_color):
        popup = tk.Toplevel(self.root)
        popup.title("Prediction Result")
        popup.geometry("400x200")
        popup.configure(bg=bg_color)
        tk.Label(popup, text="Result", font=("Helvetica", 18, "bold"), bg=bg_color, fg="white").pack(pady=10)
        tk.Label(popup, text=message, font=("Arial", 14), wraplength=350, bg=bg_color, fg="white", justify="center").pack(pady=10)
        tk.Button(popup, text="OK", command=popup.destroy, bg="white", font=("Arial", 12)).pack(pady=10)

if __name__ == '__main__':
    root = tk.Tk()
    app = FakeNewsApp(root)
    root.mainloop()
```
